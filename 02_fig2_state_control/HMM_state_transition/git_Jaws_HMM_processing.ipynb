# HMM analysis for pre/stim neural-state classification.
Loads CSV files from git_input_data/ and saves MATLAB-compatible results to git_output_data/.

```python
import numpy as np
import pandas as pd
from pathlib import Path
from scipy.io import savemat
import pickle
from data_mkv_hmm_plot import hmm_process


def process_csv_file(csv_file, output_folder_path):
    try:
        df = pd.read_csv(csv_file, delimiter=",")
        file_name = csv_file.stem

        emission_df_normalized, \
        emission_df_with_markers, \
        transmat_condition1, \
        transmat_condition2, \
        frobenius_difference, \
        transmat_condition1_ordered, \
        transmat_condition2_ordered, \
        transmat_all = hmm_process(df, file_name, output_folder_path)
                
        
        result_filename = output_folder_path / file_name / "HMM_result.mat"

        # Create a dictionary with variable names and their corresponding values
        variables_to_save = {
            'emission_df_normalized': emission_df_normalized,
            'emission_df_with_markers': emission_df_with_markers,
            'transmat_condition1': transmat_condition1,
            'transmat_condition2': transmat_condition2,
            'frobenius_difference': frobenius_difference,
            'transmat_condition1_ordered': transmat_condition1_ordered,
            'transmat_condition2_ordered': transmat_condition2_ordered,
            'transmat_all': transmat_all
        }

        # Save the variables to the MATLAB file
        savemat(result_filename, variables_to_save)

    except Exception as e:
        print(f"Error processing file {csv_file}: {e}")


input_folder_path = Path("git_input_data")
output_folder_path = Path("git_output_data")

# Use glob to find all CSV files in the input folder
csv_files = input_folder_path.glob("*.csv")

# Loop through each CSV file and process it
for csv_file in csv_files:
    process_csv_file(csv_file, output_folder_path)
    # clear_workspace()  # Clear variables at the end of each iteration
```
